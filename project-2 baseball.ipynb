```python
from  sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
X=df.drop(['W'],axis=1)
y=df['W']
```

```python
feature= SelectKBest(score_func=chi2,k=16)
feature_score=feature. fit(X,y)
```

```python
scores=feature_score.scores_
```

```python
scores
```

```python
best=pd.DataFrame(scores,columns=['score'])
best
```

```python
fea=pd.DataFrame(X.columns,columns=['feature'])
```

```python
best_score=pd.concat([fea,best],axis=1)
```

```python
best_score
```

```python
best_score.nlargest(16,'score')
```

```python
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
x=std.fit_transform(X)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin=LinearRegression()
lin.fit(x,y)
lin.score(x,y)
```

```python
y_pred=lin.predict(x)
```

```python
from sklearn import metrics
print("Mae",metrics.mean_absolute_error(y,y_pred))
print("MSE",metrics.mean_squared_error(y,y_pred))
print('RMSE',np.sqrt(metrics.mean_squared_error(y,y_pred)))
```

```python
y_predict=pd.DataFrame(y_pred,columns=['Wins'])
y_predict.to_csv("submissiondt-Wins project.csv")
```
